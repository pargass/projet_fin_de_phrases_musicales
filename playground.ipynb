```python
import MTCFeatures
from MTCFeatures import MTCFeatureLoader
import pandas as pd
```

```python
# Run once
MTCFeatures.downloadData(dest='user')
```

```python
fl = MTCFeatureLoader('MTC-FS-INST-2.0')
seqs = fl.sequences()
```

```python
type(seqs)
```

```python
item = seqs.__next__()
print(item)
print(item.keys())
print(item['features'].keys())
```

```python
%%time
phrase_data = []
for ii, x in enumerate(seqs):
    phrase_data.append({
        'id': x['id'],
        **x['features']
    })
```

```python
df_phrase = pd.DataFrame(phrase_data)
df_phrase
```
